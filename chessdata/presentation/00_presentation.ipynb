## Признаки

**Ошибки**

- Количество, процент
- Ошибки за первые несколько ходов
- Средний номер хода с ошибками
- Ошибки фигурами
    
**Eval**

- Среднее значение за партию
- Количество почти равных позиций
- Количество безнадежных позиций
- Квантили

**Потери преимущества**

- Средняя потеря преимущества
- Потеря в первые несколько ходов
- В разрезе фигур
- Дисперсия

**Особые ходы**

- Шахи, взятия, рокировки
- Количество, процент
- Качество особых ходов
    
**Неизбежные маты**

- Количество, процент
- Утеряно матов
    
**Дебют**

- Mean Encoding для названия
- Категория
